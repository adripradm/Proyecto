```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor as RFR
%matplotlib inline
import pylab as plt
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler

import h2o
from h2o.automl import H2OAutoML
scaler=StandardScaler()

from sklearn.metrics import mean_squared_error as mse
```

```python
data = pd.read_csv('data.csv', sep=';',on_bad_lines='skip')
```

```python
data.price=data.price
```

```python
data.head()
```

```python
data['year']=np.floor(2018-data.months_old/12)
```

```python
data['year'] = pd.to_datetime(data['year'], format='%Y').dt.year
```

```python
data.num_owners.fillna('Unknown', inplace=True)
data.dropna(axis=0, inplace=True)
```

```python
data.info()
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(data.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)



sns.heatmap(data.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5},
           annot=True);

```

```python
data.columns
columnas=['make', 'model', 'version', 'months_old', 'power', 'sale_type',
       'num_owners', 'gear_type', 'fuel_type', 'kms', 'year']
columnae=['make', 'model', 'version', 'months_old', 'power', 'sale_type',
       'num_owners', 'gear_type', 'fuel_type', 'kms', 'year', 'price']
entreno=data.copy()
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(data[columnae].corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)



sns.heatmap(data[columnae].corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5},
           annot=True);
```

```python
X=entreno.drop('price', axis=1)

y=entreno.price
```

```python
entreno.model.unique()
```

```python
h2o.init()
```

```python
automl=H2OAutoML(max_models=50,
                 seed=42,   # random_state
                 max_runtime_secs=300,
                 sort_metric='RMSE')

```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, train_size=0.8, random_state=23)
```

```python
h2train=h2o.H2OFrame(entreno[:7288])
h2test=h2o.H2OFrame(X[-1822:])
```

```python
X_h=[c for c in h2train.columns if c!='price']
y_h='price'

```

```python
automl.train(x=X_h,
             y=y_h,
             training_frame=h2train)

 
```

```python
print('[INFO] Leader board:')

leader_board=automl.leaderboard

leader_board.head()

```

```python
y_pred=automl.leader.predict(h2test)

```

```python
pred= h2o.as_list(y_pred)
```

```python
from sklearn.metrics import mean_squared_error as mse

mse(pred, entreno.price[-1822:], squared=False)
```

```python
from sklearn.metrics import r2_score as r2

r2(pred, entreno.price[-1822:])

```
